# Základní datové typy a proměnné


## Práce s textem "řetězce"

Pro zápis textu musíme vždy použít uvozovky **(`"`)** neho apostrof **(`'`)**

### Funkce `print()`
"Vypíše vše, co jí řekneme"

```python
print('Guns of the patriots')
print('Guns', 'of the', 'patriots')
print('Guns \nof the patriots')
```

### Spojování textu

```python
print('Guns' 'Guns' 'Guns')
print('@'*10)
```

### Manipulace s textem

```python
print('Guns of the patriots'.lower())
print('Guns of the patriots'.upper())
print('Guns of the patriots'.replace('o','T'))
print('Guns of the patriots'.replace(' ','').lower())
```

### Podřetězce

```python
print(len('Guns of the patriots'))
print('prvni znak'[0])
print('sedmy znak'[6])
print('prvnich pet znaku'[0:5])
print('poslednich pet znaku'[-5:])
print('od pateho do konce'[4:])
```

### Základní operace s čísly

```python
print('1 + 2 =',1+2)
print('1 + 2 =','1'+'2')
print('1 + 2 =',1+int('2'))
print('1 + 2 =',1+float('2'))
```

## Proměnné
**"Proměnná"** je jméno, za kterým se skrývá nějaká hodnota   
Hotnotu získává přiřazovacím příkazem `jmeno_promenne = vyraz`   
Její hodnotu získáme uvedením jejího jména ve výrazu nebo na jiném vhodném místě   
Jméno proměnné může obsahovat alfanumerické znaky a podtržítko (`_`), nesmí začínat číslem   
Doporučuje se používat vhodně popisná jména; ideálně vycházející z angličtiny

```python
a = 4
print(a)

b = 45
print(b)

print('a + b =',a+b)

print(f'{a:.2f} + {b} =',a+b)

ROVNICE = 'nase rovnice: {} + {} = {}'
print(ROVNICE.format(a,b,a+b))
print(ROVNICE.format(3,45,44))
```
