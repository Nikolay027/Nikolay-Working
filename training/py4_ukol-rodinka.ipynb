# Úkol: Rodinka

Navrhněte datovou strukturu, ve které popíšete rodinné vztahy:
- každá osoba bude v proměnné typu `dict` a musí mít jméno `jmeno`, další informace jsou volitelné
- minimálně tři generace
- v každé generaci minimálně dvě osoby
- rodiče jsou vzájemně propojeni
- rodiče mají vazbu na své potomky

## Import

```python
import yaml
```

## Definice konstant


```python
KEY_NAME = 'jmeno'
KEY_PARTNER = 'partner'
KEY_CHILDREN = 'deti'
KEY_EXTRAS = 'zajimavost'
```

## Datová struktura

osoby /uložené v proměnných/

`rodina =` seznam všech osob

```python
Valeriy = {KEY_NAME:'Valeriy Spitsyn',KEY_PARTNER:'Galina',KEY_CHILDREN:'Nikolay', KEY_EXTRAS: 'Metalurg'}
Galina = {KEY_NAME:'Galina Spitsyna',KEY_PARTNER:'Valeriy',KEY_CHILDREN:'Nikolay', KEY_EXTRAS: 'Programatorka'}
Nikolay = {KEY_NAME:'Nikolay Spitsy',KEY_PARTNER:'NONE',KEY_CHILDREN:'NONE', KEY_EXTRAS: 'Gamer'}
Rodina1 = [Valeriy, Galina, Nikolay]
```

vazby ... definice vazeb mezi členy rodiny

```python
print(yaml.dump(Rodina1))
```
