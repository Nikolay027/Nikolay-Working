# Úkol: Rodinka

Navrhněte datovou strukturu, ve které popíšete rodinné vztahy:
- každá osoba bude v proměnné typu `dict` a musí mít jméno `jmeno`, další informace jsou volitelné
- minimálně tři generace
- v každé generaci minimálně dvě osoby
- rodiče jsou vzájemně propojeni
- rodiče mají vazbu na své potomky

## Import

```python
import yaml
```

## Definice konstant


```python
KEY_NAME = 'jmeno'
KEY_PARTNER = 'partner'
KEY_CHILDREN = 'deti'
KEY_EXTRAS = 'zajimavost'
```

## Datová struktura

osoby /uložené v proměnných/

`rodina =` seznam všech osob

```python
Valeriy = {KEY_NAME:'Valeriy Spitsyn',KEY_PARTNER:'Galina',KEY_CHILDREN:'Nikolay', KEY_EXTRAS: 'Metalurg'}
Galina = {KEY_NAME:'Galina Spitsyna',KEY_PARTNER:'Valeriy',KEY_CHILDREN:'Nikolay', KEY_EXTRAS: 'Programatorka'}
Nikolay = {KEY_NAME:'Nikolay Spitsy',KEY_PARTNER:'NONE',KEY_CHILDREN:' ', KEY_EXTRAS: 'Gamer'}
Rodina1 = [Valeriy, Galina, Nikolay]
Valeriy[KEY_PARTNER] = Galina
Galina[KEY_PARTNER] = Valeriy
Nikolay[KEY_PARTNER] = None
Valeriy[KEY_CHILDREN] = [Nikolay]
Galina[KEY_CHILDREN] = [Nikolay]
Nikolay[KEY_CHILDREN] = []

```

vazby ... definice vazeb mezi členy rodiny

```python
print(yaml.dump(Rodina1))
```

## Práce s rodinkou
cyklus `for`  
Podminka `if`

```python
for osoba in Rodina1:
    print(osoba.get(KEY_NAME),end = ' ')
    print(f'ma partnera: {osoba.get(KEY_PARTNER,{}).get(KEY_NAME)}'if osoba.get(KEY_PARTNER,) else 'partnera nema')
    if osoba.get(KEY_CHILDREN):
        print('\tma deti:',osoba.get(KEY_CHILDREN))
    else:
        print('\tdeti nema')
```

```python
[child.get(KEY_NAME) for child in Valeriy.get(KEY_CHILDREN)]
```
