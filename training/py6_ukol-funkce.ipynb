# Úkol: vytvoř funkci

Tvým úkolem je doplnit funkční kód do připravených funkcí:
- `na_kostce()` vrátí pojmenování hodnoty na kostce
- `score()` vypočítá skóre hodu 6 kostkami a vrátí výsledek
- `calculate()` provede zvolenou matematickou operaci
   nad čísly předanými jako parametr a vrátí výsledek


## Hodnota na kostce

```python
def na_kostce(val):
    '''
    Pokud je hodnota parametru `val` '1' až '6', vrátí funkce jméno číslice,
    v každém jiném připadě vrátí hodnotu s komentářem, že nic takového na kostce není.
    '''
    output = None
    ########## prostor pro splnění úkolu #########
    output = f"{val} na kostce není"
    if val == 1:
        output = "jednička"
    elif val == 2:
        output = "dvojka"
    elif val == 3:
        output = "trojka"
    elif val == 4:
        output = "čtyři"
    elif val == 5:
        output = "pět"
    elif val == 6:
        output = "šest"


    
    ###########################################
    return output

# cyklus pro otestovani funkce
for ii in range(-2,11):
    print(ii,' ... ',na_kostce(ii))
```

## Skóre postupky

```python
def score(*throw):
    '''
    vypočítá skóre hodu 6 kostkami a vrátí výsledek
    Výpočet má následující pravidla:
    1. hodnotí se jen postupky začínající jedničkou /ostatní čísla nehrají roli/
    2. body za postupky jsou následující:
    | Postupka     | Body |
    |--------------|------|
    | 1-2          |  20  |
    | 1-2-3        |  30  |
    | 1-2-3-4      |  50  |
    | 1-2-3-4-5    |  90  |
    | 1-2-3-4-5-6  |  200  |
    
    Parameters
    ----------
    throw: neuspořádaný seznam hodnot kostek 
    
    Returns
    -------
    numeric
    
    '''
    
    score = 0
    ########## prostor pro splnění úkolu #########
    if len(throw) == 6:
        if 1 in throw:
            if 2 in throw:
                score = 20
                if 3 in throw:
                    score = 30
                    if 4 in throw:
                        score = 50
                        if 5 in throw:
                            score = 90
                            if 6 in throw:
                                score = 200

    
    
    ##############################################
    return score


# ověření správnosti funkce
TESTS = (
    ([1,2,3,4,5,6],200),
    ([6,6,3,2,1,7],30),
    ([1,2,4,4,5,6],20),
    ([2,2,3,4,1,6],50),
    ([2,5,3,4,1,3],90),
    ([1,2,3,4],0),
    ([6,5,4,4,3,2,1],0),
)
ok = True
for test in TESTS:
    res = score(*test[0])
    tok = res == test[1]
    if tok:
        print(test[0]," - OK")
    else:
        print(test[0],f" - expected '{test[1]}', got '{res}'")
    ok = ok & tok
print("Celkový výsledek:","OK" if ok else "někde je chyba")
```

## Kalkulačka

povolené operace jsou v `class OPERATION`

```python
class OPERATION:
    ''' seznam operací, se kterými pracujeme
    Konkrétní operaci získáš zápisem `OPERATIONS.<jmeno-operace>`
    Příklad: `if OPERATION.ADD == 'scitani':` 
    '''
    ADD = 'scitani'
    ''' kód pro operaci sčítání '''
    SUB = 'odcitani'
    ''' kód pro operaci odčítání '''
    MULTI = 'nasobeni'
    ''' kód pro operaci násobení '''
    DIV = 'deleni'
    ''' kód pro operaci dělení '''

def calculate(*nums,todo:OPERATION):  #### zde doplň správnou definici parametrů
    '''
    funkce provede zadanou operaci nad libovolným počtem čísel
    
    Parameters
    ----------
    nums: libovolný počet čísel
    todo: kód požadované operace (pojmenovaný parametr)
    
    Returns
    -------
    numeric
    
    '''
    
    result = 0
    ########## prostor pro splnění úkolu #########
    result= nums[0] if len(nums) > 0 else 0
    if todo == OPERATION.ADD:
        for ii in range(1,len(nums)):
            result += nums[ii]
    elif todo == OPERATION.SUB:
        for ii in range(1,len(nums)):
            result -= nums[ii]
    elif todo == OPERATION.MULTI:
        for ii in range(1,len(nums)):
            result *= nums[ii]
    elif todo == OPERATION.DIV:
        for ii in range(1,len(nums)):
            result /= nums[ii]

    ##############################################
    return result


# ověření správnosti funkce
TESTS = (
    ([1,2,3,4,5,6],21,OPERATION.ADD),
    ([6,6,3,2,1,7],-13,OPERATION.SUB),
    ([12,3,4,1,2],0.5,OPERATION.DIV),
    ([1,2,3,4],24,OPERATION.MULTI),
    ([1],1,OPERATION.MULTI),
    ([],0,OPERATION.ADD),
    ([],0,OPERATION.SUB),
    ([],0,OPERATION.MULTI),
    ([],0,OPERATION.DIV),
)
ok = True
for test in TESTS:
    res = calculate(*test[0],todo=test[2])
    tok = res == test[1]
    if tok:
        print(f"{test[2]} {test[0]}: - OK")
    else:
        print(f"{test[2]} {test[0]}: ERROR - expected '{test[1]}', got '{res}'")
    ok = ok & tok
print("Celkový výsledek:","OK" if ok else "někde je chyba")
```
