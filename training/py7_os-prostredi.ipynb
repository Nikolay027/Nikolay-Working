## Zápis `dict` na disk
Zkopíruj si rodinku z [`py4_ukol-rodinka.ipynb`](py4_ukol-rodinka.ipynb)

```python
KEY_NAME = 'jmeno'
KEY_PARTNER = 'partner'
KEY_CHILDREN = 'deti'
KEY_EXTRAS = 'zajimavost'

Valeriy = {KEY_NAME:'Valeriy Spitsyn',KEY_PARTNER:'Galina',KEY_CHILDREN:'Nikolay', KEY_EXTRAS: 'Metalurg'}
Galina = {KEY_NAME:'Galina Spitsyna',KEY_PARTNER:'Valeriy',KEY_CHILDREN:'Nikolay', KEY_EXTRAS: 'Programatorka'}
Nikolay = {KEY_NAME:'Nikolay Spitsy',KEY_PARTNER:'NONE',KEY_CHILDREN:' ', KEY_EXTRAS: 'Gamer'}
Rodina1 = [Valeriy, Galina, Nikolay]
Valeriy[KEY_PARTNER] = Galina
Galina[KEY_PARTNER] = Valeriy
Nikolay[KEY_PARTNER] = None
Valeriy[KEY_CHILDREN] = [Nikolay]
Galina[KEY_CHILDREN] = [Nikolay]
Nikolay[KEY_CHILDREN] = []

soubor = os.path.join(WORK_FOLDER,'rodina.yml')
with open(soubor,'w',encoding='UTF-8') as ff:
    yaml.dump(Rodina1,ff)
```

#### Rodinku přečteme zpět

```python
rodina2 = None
with open(soubor,'r',encoding='UTF-8') as ff:
    rodina2 = yaml.safe_load(ff)
print(type(rodina2),len(rodina2))
print(rodina2[2])
```
